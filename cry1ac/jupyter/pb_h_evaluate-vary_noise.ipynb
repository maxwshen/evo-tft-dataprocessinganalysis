```python
import numpy as np, matplotlib.pyplot as plt, pandas as pd, seaborn as sns

import sys
sys.path.append('/ahg/regevdata/projects/CRISPR-libraries/prj2/evolution/badran/src/')
import _config

import pb_h_evaluate_exps as parent_script
```

```python
modelexp_nm = 'modelexp_varynoise_rs'

df = pd.read_csv(parent_script.out_dir + f'{modelexp_nm}.csv', index_col = 0)

print(df.columns)
```

## plots

```python
fig, ax = plt.subplots()

sns.boxplot(
    x = 'data_noise',
    y = 'Pearsonr (overall)',
    data = df,
    color = 'white',
    fliersize = 0,
    ax = ax,
)
sns.stripplot(
    x = 'data_noise',
    y = 'Pearsonr (overall)',
    data = df,
    alpha = 0.4,
    ax = ax,
)

sns.despine()
plt.grid(alpha = 0.2)

plt.xlabel('Synthetic noise');
plt.ylabel('Performance (Pearson correlation)');
plt.xticks(rotation = 90);
```
