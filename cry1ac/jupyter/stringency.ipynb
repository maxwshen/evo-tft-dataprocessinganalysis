```python
import numpy as np, matplotlib.pyplot as plt, pandas as pd, seaborn as sns
```

```python
def func(x, n):
    return 1 - (1-x)**n

ns = [1, 2, 5, 10, 50]
for n in ns:
    xs = np.arange(0, 1, 0.005)
    ys = [func(x, n) for x in xs]

    plt.plot(
        xs, 
        ys,
        label = f'{n} APs per E. coli cell',
    )

sns.despine()
plt.grid(alpha = 0.2)
plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left')
plt.xlabel('Genotype fitness (probability of activating gene III expression)');
plt.ylabel('Probability of generating gene III among N APs\n -> phage propagation -> fitness');
```
