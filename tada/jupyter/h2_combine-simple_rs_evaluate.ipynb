```python
import numpy as np, seaborn as sns, matplotlib.pyplot as plt, pandas as pd

import sys
sys.path.append('/ahg/regevdata/projects/CRISPR-libraries/prj2/evolution/abe8e/src/')
import h2_combine as parent_script
import _config
```

```python
df = pd.read_csv(parent_script.out_dir + f'modelexp_simple_rs.csv', index_col = 0)
print(df.shape)

print(df.head())
```

## exploratory

```python
target_col = 'TimewiseR 25%'
```

```python
sns.violinplot(
    x = 'pace_num',
    y = target_col,
    hue = 'threshold',
    data = df,
)
sns.despine()
plt.grid(alpha = 0.2)
plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left')
```

```python
sns.violinplot(
    x = 'pace_num',
    y = target_col,
    hue = 'min_gt_frequency',
    data = df,
)
sns.despine()
plt.grid(alpha = 0.2)
plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left')
```

```python
sns.violinplot(
    x = 'read_len',
    y = target_col,
    data = df,
)
sns.despine()
plt.grid(alpha = 0.2)
plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left')
```
